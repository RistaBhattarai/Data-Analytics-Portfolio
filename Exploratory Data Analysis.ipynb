```python
# Understanding prostate cancer using subplots

f, axes = plt.subplots(1, 2, figsize=(18,5)) # creates subplot grid with 1 row and 2 columns
sns.distplot(df['Prostate Cancer'], ax=axes[0])
sns.boxplot(df['Prostate Cancer'], ax=axes[1])
```

```python
# Understanding different behaviours of an adult

f, axes = plt.subplots(2, 2, figsize=(18,10)) # creates subplot grid with 2 rows and 2 columns
sns.distplot(df['Seatbelt Use'], ax=axes[0][0])
sns.boxplot(df['Immunization'], ax=axes[0][1])
sns.scatterplot(x="Health Care Coverage", y="Pneumonia Shot Ever", data=df, ax=axes[1][0])
sns.scatterplot(x="Lifetime Asthma", y="Smoking", data=df, ax=axes[1][1], hue="Sex")
```

```python
# Understanding physical and mental health

f, axes = plt.subplots(2, 1, figsize=(10,6), sharex=True) # creates subplot grid with 2 row and 1 columns
sns.boxplot(df['Physical Health'], ax=axes[0])
sns.distplot(df['Mental Health'], ax=axes[1])
```

```python
# Understanding about heart disease among adults

sns.relplot(x="Heart Attack", y="Angina or Coronary heart Disease", data=df, 
            hue="Stroke",
            col="COPD" # categorical variable according to which the columns of the grid will be determined
           )
```

```python
# Understanding multiple variables relationship

sns.catplot(x="Kidney Disease", y="Alcoholic Beverage", data=df, kind="bar",
            col="Sex",
            row="COPD",
            hue="Stroke",
            ci=None
           )
```

```python
# Relationship between general, physical and mental health

sns.pairplot(df, vars=['General Health', 'Physical Health', 'Mental Health'],
             hue="Sex"                
            )
```
