# Checking for null values

```python
pd.set_option('display.max_rows', 200)

df.isnull().sum(axis = 0)
```

```python
df[df.columns[df.isnull().any()]]
```

```python
df_new = df
```

# Filling the missing values 

```python
# Filling the missing values using interpolate method

df_new = df_new.interpolate(method ='linear', limit_direction ='forward')
```

```python
df_new.isnull().sum(axis = 0)
```

# Visualizing the variables 

```python
# Visualizing the variables before and after the values are filled

df.plot(kind = 'scatter', x = 'Poor Health', y = 'State')

plt.show() 
```

```python
fig, ax = plt.subplots()
ax2 = ax.twinx()

df.plot(x="State", y= "Poor Health", label = "Poor Health with null values", ax=ax)
df_new.plot(x="State", y= "Poor Health", label = "Poor Health without null values", ax=ax2, ls="--")

plt.show()
```
